## Peças comprimidas
    
</font>

**Além da verificação de estabilidade** de, acordo com o item 6.5 da ABNT NBR 7190-1:2022, a condição de segurança relativa à resistência à compressão axial é calculada conforme a seguinte equação:

----  
    
$
\sigma N_{c,d} = \frac{N_{c,d}}{A}\leq f_{c0,d}
$
    
------

    
<p>
onde:
<p>
- $\sigma N_{c,d}$:  é o valor de cálculo da tensão de compressão normal à seção transversal;      
<p>      
- $N_{c,d}$: é o valor de cálculo da tensão de compressão normal à seção transversal;
<p>       
 - $A$: Área líquida da seção transversal;
<p>  
- $f_{c0,d}$: é o valor de cálculo da resistência à compressão paralela às fibras.
<p>
    
> No caso de peças com fibras inclinadas de ângulos $\alpha > 6^{o}$, aplica-se à $f_{c0,d}$ a redução mostrada no item 6.2.8 da ABNT NBR 7190-1:2022.

> Para madeira lamelada colada cruzada, a área da seção transversal deve ser calculada conforme item 6.7.4.10.2 da ABNT NBR 7190-1:2022.

## Estabilidade

</font>
 
Segundo item 6.5 da ABNT NBR 7190-1:2022

O procedimento para verificação da estabilidde de peças comprimidas deve atender a:

### 1 .  Condições de alinhamento da peça

Para peças que compõem pórticos, treliças, pilares ou vigas em que a instabilidade lateral pode ocorrer, o desvio no alinhamento axial da peça, medido na metade da distância entre os apoios, deve ser limitado em:

  - $ \frac{L}{300}$ para peças de madeira serrada ou roliça;
  
  - $ \frac{L}{500}$ para peças de madeira laminada colada.
  
### 2 .  Esbeltez

 - Índice de esbeltez($\lambda$): 
    
>> O índice de esbeltez das peças sujeitas à compressão axial ou à flexocompressão não pode ser maior que 140.
    
    
 $
\lambda = \frac{L_{0}}{\sqrt{\frac{I}{A}}}
 $
 
 $L_{0}$: Comprimento de flambagem;
 $I$: Momento de inércia;
 $A$: Área.
 
 $L_{0} = k_{E}L$
 
 $k_{E}$: Coeficiente de flambagem. Leva em conta o tipo de vinculação da peça, como mostra a tabela:
 


### 3. Esbeltez relativa    
    
- Na direção $x$:   
   
 $
\lambda_{rel_{x}} = \frac{\lambda_{x}}{\pi}\sqrt{\frac{f_{c0,k}}{E_{0.05}}}
 $
    
- Na direção $y$:
    
 $
\lambda_{rel_{y}} = \frac{\lambda_{y}}{\pi}\sqrt{\frac{f_{c0,k}}{E_{0.05}}}
 $   
    
    
$E_{0.05} = 0.7 E_{c0,med}$. Ítem 5.8.7 da ABNT NBR 7190-1:2022
    

    
### 4 - Condição de estabilidade de peças comprimidas e flexocomprimidas  
    
- Para $\lambda_{rel_{x}} < 0,3 $ **e** $\lambda_{rel_{y}} < 0,3 $:
    
Não são necessárias as verificações deste item.
    
 <p>   
    
- Caso contrário, devem ser verificadas as inequações:
  <p>   
      <br>
     $
    \frac{\sigma_{N_{c,d}}}{k_{cx}f_{c0,d}} +
    \frac{\sigma_{M_{x,d}}}{f_{m,d}} + 
    k_{M}\frac{\sigma_{M_{y,d}}}{f_{m,d}} \leq 1
    $ e
      
   <p>
       <br>
     $
    \frac{\sigma_{N_{c,d}}}{k_{cy}f_{c0,d}} +
    \frac{\sigma_{M_{x,d}}}{f_{m,d}} + 
    k_{M}\frac{\sigma_{M_{y,d}}}{f_{m,d}} \leq 1
    $   
      
       
 <p> Para secão retangulares $k_{M} = 0,4$, nas outras, $k_{M} = 1$.
       
 $\sigma_{M}$ é a tensão normal de flexão proveniente do momento fletor de primeira ordem devida às forças laterais, excentricidades na aplicação das forças axiais, curvatura inicial da barra, deformações induzidas ou quaisquer outras situações em que há momentos fletores de primeira ordem atuando na barra;
<p>
$k_{cx} = \frac{1}{k_{x} +  \sqrt{k_{x}^{2} - \lambda_{rel_{x}}^{2} }} $
<p>
$k_{cy} = \frac{1}{k_{y} +  \sqrt{k_{y}^{2} - \lambda_{rel_{y}}^{2} }} $
 <p>   
$ k_{x} = 0,5 \left[ 1+ \beta_{c}(\lambda_{rel_{x}} - 0,3) + \lambda_{rel_{x}}^{2} \right]  $

$ k_{y} = 0,5 \left[ 1+ \beta_{c}(\lambda_{rel_{y}} - 0,3) + \lambda_{rel_{y}}^{2} \right]  $

 Para madeiras serradas e roliças em peças estruturais que atendam aos limites de divergência de alinhamento, $ \beta_{c} = 0,2 $

## Exemplo
    
</font>     

Verificar o banzo comprimido de treliça.

- L0 = 169 cm

- Madeira: Dicotiledônea – classe D 60

- Local de classe de umidade 1

- Seção transversal $6 cm x 16 cm$

- solicitação:
    - Carga permanente = $2.400 daN = 24kN $
    - Vento de pressão = $564 daN$
    
    >> $daN$: decanewton, $1daN = 10N$.


<font size="+1" face="Arial" color="blue">

### Solução
</font> 

 #### 1 - Propriedades geométricas:

```python
b = 6 #[cm]
h = 16 # [cm]
A = b * h # Área em cm²
Ix = b * (h**3) / 12 # maior momento de inèrcia em cm⁴
Iy = h * (b**3) /12 # menor momento de inèrcia em cm⁴
rx = (Ix / A)**0.5
ry = (Iy / A)**0.5
```

#### 2 - Propriedades Físicas:
    
 Madeira: Classe D 60
 
 $ f_{c0,k} = 60 MPa$
 
 $ f_{c0,d} = K_{mod} \frac{ f_{c0,k}}{\gamma_{w}}$
  

```python
f_c0k = 60 #[MPa]  Tabela 2 da ABNT NBR 7190-1:2022
f_c0d = 0.70 * f_c0k  / 1.4 # [MPa]
print("𝑓𝑐0,𝑑 = ", f_c0d, "MPa")
```

ABNT NBR 7190-1:2022, item 5.8.7:
>Nas verificações de estados limites últimos referentes à estabilidade de peças comprimidas e flexocomprimidas, deve ser utilizado o valor característico para o módulo de elasticidade, $E_{0,05}$.
>No caso do uso da Tabela 2 o valor característico pode ser utilizado como sendo igual a 70 % do valor médio do módulo de elasticidade:

$
E_{0,05} = 0, 7 E_{c0,med}
$


```python
E_c0med = 19500 #MPa - Tabela 2 da ABNT NBR 7190-1:2022 
E_05 = 0.7 * E_c0med 
E_05
```

#### 3 - Esbeltez

 $
\lambda = \frac{L_{0}}{\sqrt{\frac{I}{A}}}
 $

```python
L_0 = 169 # cm
lbda_y = L_0 / ry # o maior
lbda_x = L_0 / rx 
print("𝜆𝑦 =",lbda_y , "𝜆x =",lbda_x ) # 
```

<font size="+1" face="Arial" color="red">
OK!! $ \lambda_x < 140$ e  $ \lambda_y < 140$!!
<font

2 - Verificando a esbeltez reletiva

- Na direção $x$:   
   
 $
\lambda_{rel_{x}} = \frac{\lambda_{x}}{\pi}\sqrt{\frac{f_{c0,k}}{E_{0.05}}}
 $
    

```python
from math import pi
lbda_rel_x = (lbda_x / pi) * (f_c0k / E_05)**0.5
print("𝜆𝑟𝑒𝑙𝑥 = ", lbda_rel_x)
```

- Na direção $y$:   
   
 $
\lambda_{rel_{y}} = \frac{\lambda_{y}}{\pi}\sqrt{\frac{f_{c0,k}}{E_{0.05}}}
 $
    

```python
lbda_rel_y = (lbda_y / pi) * (f_c0k / E_05)**0.5
print("𝜆𝑟𝑒𝑙y = ", lbda_rel_y)
```

Como <font size="+1" face="Arial" color="red">$\lambda_{rel_{x}} > 0.3 $ **e** $\lambda_{rel_{y}} > 0.3 $</font>, precisamos verificar:           


 
 $
    \frac{\sigma_{N_{c,d}}}{k_{cx}f_{c0,d}} +
    \frac{\sigma_{M_{x,d}}}{f_{m,d}} + 
    k_{M}\frac{\sigma_{M_{y,d}}}{f_{m,d}} \leq 1
    $ e
      
   <p>
       <br>
     $
    \frac{\sigma_{N_{c,d}}}{k_{cy}f_{c0,d}} +
    \frac{\sigma_{M_{x,d}}}{f_{m,d}} + 
    k_{M}\frac{\sigma_{M_{y,d}}}{f_{m,d}} \leq 1
    $   
    

Observemos que:

$\sigma_{M_{x,d}} = 0$ 

$\sigma_{M_{y,d}} = 0$

$\sigma_{N_{c,d}} = \frac{F_{d}}{A}$

#### 4 - Tensão de cálculo, $\sigma_{N_{c,d}}$.

De acordo com a ABNT NBR 8681:2003, considerando a ABNT NBR 7190-1:2023:



- Carga permanente: $F_{g1} = 24kN$
- Vento de pressão, carga variável: $F_{q1} = 5.64 kN$
- $\gamma_{g1} = 1.30$
- $\gamma_{q} = 1.40$ e deve ser multiplicado por $0.75$

```python
F_d = 1.30 * 24 + 0.75 * 1.40 * 5.64 #kN
print("𝐹𝑑 =", F_d, "kN")
```

$\sigma_{N_{c,d}} = \frac{F_{d}}{A}$ 

```python
Sigma_ncd = (F_d * 1e3) / (A * 1e-4) #[Pa]
# [Pa] ---> [MPa]
Sigma_ncd = Sigma_ncd * 1e-6 #[Mpa]
print("𝜎𝑁𝑐,𝑑 =", Sigma_ncd, "MPa")
```

<p>
$k_{cx} = \frac{1}{k_{x} +  \sqrt{k_{x}^{2} - \lambda_{rel_{x}}^{2} }} $
<p>
$k_{cy} = \frac{1}{k_{y} +  \sqrt{k_{y}^{2} - \lambda_{rel_{y}}^{2} }} $
 <p>   
$ k_{x} = 0,5 \left[ 1+ \beta_{c}(\lambda_{rel_{x}} - 0,3) + \lambda_{rel_{x}}^{2} \right]  $

$ k_{y} = 0,5 \left[ 1+ \beta_{c}(\lambda_{rel_{y}} - 0,3) + \lambda_{rel_{y}}^{2} \right]  $

```python
k_x = 0.5 * (1 + 0.2 * (lbda_rel_x - 0.3) + lbda_rel_x ** 2)
k_y = 0.5 * (1 + 0.2 * (lbda_rel_y - 0.3) + lbda_rel_y ** 2)
k_cx = 1 / (k_x + (k_x ** 2 + lbda_rel_x ** 2) ** 0.5)
k_cy = 1 / (k_y + (k_y ** 2 + lbda_rel_y ** 2) ** 0.5)
```

```python
print(Sigma_ncd / (k_cx * f_c0d))
```

 $ \frac{\sigma_{N_{c,d}}}{k_{cx}f_{c0,d}} < 1$ <font size="+1" face="Arial" color="red"> OK!!</font>

```python
print(Sigma_ncd / (k_cy * f_c0d))
```

$ \frac{\sigma_{N_{c,d}}}{k_{cy}f_{c0,d}} < 1$ <font size="+1" face="Arial" color="red">OK!!</font>

#### Conclusão:

Peça <font size="+1" face="Arial" color="red">OK!!</font>
